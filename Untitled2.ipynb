```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
```

```python
# STEP 2: Load dataset

df = pd.read_csv("ultimate_student_productivity_dataset_5000.csv")
df.head()
```

```python
# STEP 3: View column names

df = df.dropna()
print("Data cleaned successfully")
```

```python
# STEP 4: Dataset information

df.info()
```

```python
# STEP 5: Check missing values

print(df.isnull().sum())

# Remove missing values

df = df.dropna()

print("Step 5 Complete: Data Cleaned")
```

```python
# STEP 6: Scatter Plot

plt.figure(figsize=(8,5))

plt.scatter(df['study_hours'], df['productivity_score'])

plt.title("Relational Plot: Study Hours vs Productivity Score")

plt.xlabel("Study Hours")

plt.ylabel("Productivity Score")

plt.show()
```

```python
# STEP 7: Histogram

plt.figure(figsize=(8,5))

plt.hist(df['study_hours'], bins=30)

plt.title("Categorical Plot: Histogram of Study Hours")

plt.xlabel("Study Hours")

plt.ylabel("Frequency")

plt.show()
```

```python
# STEP 8: Statistical Plot (HEATMAP)

plt.figure(figsize=(10,8))

correlation = df.corr(numeric_only=True)

sns.heatmap(correlation, annot=True)

plt.title("Statistical Plot: Correlation Heatmap")

plt.show()
```

```python
# STEP 9: Statistical Moments

print("Mean")
print(df.mean(numeric_only=True))
```
